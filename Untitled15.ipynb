```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error

data = pd.read_csv("/content/Employee_Salary_Dataset.csv")

X = data[["Experience_Years", "Age"]]
y = data["Salary"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

lr = LinearRegression().fit(X_train, y_train)
print("Linear Train MSE:", mean_squared_error(y_train, lr.predict(X_train)))
print("Linear Test MSE:", mean_squared_error(y_test, lr.predict(X_test)))

for a in [0.01, 0.1, 1, 10]:
    ridge = Ridge(alpha=a).fit(X_train, y_train)
    print(f"Ridge (alpha={a}) Train MSE:", mean_squared_error(y_train, ridge.predict(X_train)))
    print(f"Ridge (alpha={a}) Test MSE:", mean_squared_error(y_test, ridge.predict(X_test)))

for a in [0.01, 0.1, 1, 10]:
    lasso = Lasso(alpha=a).fit(X_train, y_train)
    print(f"Lasso (alpha={a}) Train MSE:", mean_squared_error(y_train, lasso.predict(X_train)))
    print(f"Lasso (alpha={a}) Test MSE:", mean_squared_error(y_test, lasso.predict(X_test)))
```
